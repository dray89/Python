```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
time_series = pd.Series(np.random.randn(1000), 
                       index=pd.date_range('1/1/2000', periods=1000))
```

```python
time_series2 = time_series.cumsum()
time_series2[0:5]
```

```python
print(time_series2.plot())
```

```python
data_frame = pd.DataFrame(np.random.randn(1000,4), index=time_series2.index, 
                         columns=['A', 'B', 'C', 'D'])
```

```python
data_frame = data_frame.cumsum()
data_frame.head()
```

```python
plt.figure()
data_frame.plot()
plt.legend(loc='best')
```

```python
data_frame.to_csv('foo.csv')
```
