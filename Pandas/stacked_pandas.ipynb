```python
import pandas as pd
import numpy as np
```

```python
data_frame = pd.DataFrame({'A': ['Cat1', 'Cat1', 'Cat1', 'Cat2',
                             'Cat3', 'Cat3', 'Cat4', 'Cat4'],
                        'B': ['one', 'one', 'two', 'three',
                             'two', 'two', 'one', 'three'],
                        'C': np.random.randn(8),
                        'D': np.random.randn(8)})
```

```python
data_frame
```

```python
#Use groupby to group data and sum the groups with sum()
data_frame.groupby('A').sum()
```

```python
#group by column A and then B, sum results
data_frame.groupby(['A', 'B']).sum()
```

```python
#Create a tuple list from zip iterating over two lists
tuples = list(zip(*[['bar', 'bar', 'baz', 'baz',
                    'foo', 'foo', 'qux', 'qux'],
                   ['one', 'two', 'one', 'two',
                   'one', 'two', 'one', 'two']]))
```

```python
#Construct a multiindex from a tuple labeling the columns first and second
index = pd.MultiIndex.from_tuples(tuples, names = ['first', 'second'])
```

```python
#Create a data_frame object containing random numpy numbers and the multiindex labeling columns A & B
data_frame = pd.DataFrame(np.random.randn(8,2), index = index, columns=['A', 'B'])
```

```python
#Create a data_frame2 object from all columns in dataframe
data_frame2 = data_frame[:4]
```

```python
data_frame2
```

```python
#Stacked columns A & B like rows
stacked = data_frame2.stack()
```

```python
stacked
```

```python
#Unstack columns
stacked.unstack()
```

```python
stacked.unstack(1)
```

```python
stacked.unstack(0)
```

```python
data_frame = pd.DataFrame({'A': ['one', 'one', 'two', 'three']*3, 
                           'B': ['A', 'B', 'C']*4,
                           'C': [ 'foo', 'foo', 'foo', 'bar', 'bar', 'bar']*2,
                           'D': np.random.randn(12),
                           'E': np.random.randn(12)
                          })
```

```python
data_frame
```

```python
#Create Columns out of the labels in column C in dataframe, using common columns A & B as row index/grouping and values in column 'D'
pd.pivot_table(data_frame, values = 'D', index = ['A', 'B'], columns=['C'])
```

```python
#Create Columns out of the labels in column C in dataframe, using common columns A & B as row index/grouping and values in column 'E'
pd.pivot_table(data_frame, values = 'E', index = ['A', 'B'], columns=['C'])
```
