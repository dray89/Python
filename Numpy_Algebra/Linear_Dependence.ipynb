```python
# Graphically representing systems of equations
# Multiple solutions to equation systems
# Linear Combinations
    # Columns = number of vector space dimensions (n directions)
    # Number of Solutions = Number of ways we can reach the point b by traveling through the number of vector space dimensions (columns)
# Underdetermined and Overdetermined Systems

import numpy as np
import matplotlib.pyplot as plt

def plotVectors(vectors, columns, alpha=1): 
    # Create Graph
    plt.figure()
    plt.axvline(x=0, color = 'grey', zorder=0)
    plt.axhline(y=0, color = 'grey', zorder=0)
    
    for i in range(len(vectors)):
        x = np.concatenate([[0,0], vectors[i]])
        # Create Arrows
        plt.quiver([x[0]],
                  [x[1]],
                  [x[2]],
                  [x[3]],
                  angles = 'xy', scale_units='xy', scale = 1, color = columns[i], 
                  alpha = alpha)
```

```python
### Infinite number of solutions 
### One eqn, one unknown

# Ax = b
# Ay = b
# Az = b
# Ax = Ay = Az
# z = ax + (1 - a)y
# Ax = Ay = A[ax + (1 - a)y] = Axa + Ay(1 - a) = ba + by(1-a) 
# Az = ba + by(1-a) = ba + b - ba = b = z

# Define x and y values
x = np.arange(-10, 10, 1)
y = 2*x + 1

# Create figure
plt.figure()

# Plot x and y coord on graph
plt.plot(x, y)

# Define x-axis value range
plt.xlim(-2, 10)

# Define y-axis value range
plt.ylim(-2, 10)

# Draw Axes
plt.axvline(x=0, color = 'Grey')
plt.axhline(y=0, color = 'Grey')

plt.show()
plt.close()
```

```python
### 2 eqns, 1 solution, one unknown

x = np.arange(-10, 10)

# Define eqns
y = 2*x + 1 
y1 = 6*x - 2

# Create figure and lines
plt.figure()
plt.plot(x, y)
plt.plot(x, y1)

# Define Axis range
plt.xlim(-2, 10)
plt.ylim(-2, 10)

# Draw Axes
plt.axvline(x=0, color = 'Grey')
plt.axhline(y=0, color = 'Grey')

plt.show()
plt.close()

# The two lines begin at separate points and cross at one point in quadrant I
```

```python
### 3 equations, No solutions to one unknown
### Define x-value
x = np.arange(-10, 10)

# Define equations
y = 2*x + 1
y1 = 6*x - 2
y2 = .1*x + 6

# Create Graph
plt.figure()

# Plot Equations
plt.plot(x, y)
plt.plot(x, y1)
plt.plot(x, y2)

# Define Range for Axes
plt.xlim(-2, 10)
plt.ylim(-2, 10)

# Draw Axes
plt.axvline(x=0, color= 'Grey')
plt.axhline(y=0, color= 'Grey')

plt.show()
plt.close()
```

```python
# The weighted sum or linear combination of two vectors 
# Span = The set of points two vectors can reach by their combination and scaling by a / b

# Define vector weights
a = 2 
b = 1 

# Start and end coordinates of the vectors
u = [0,0,1,3]
v = [2,6,2,1]

# Define Arrows
plt.quiver([u[0], a*u[0], b*v[0]],
           [u[1], a*u[1], b*v[1]],
           [u[2], a*u[2], b*v[2]],
           [u[3], a*u[3], b*v[3]],
           angles='xy', scale_units='xy', scale=1, color=['Orange', 'Orange', 'Blue'])

# Label Axes
plt.xlim(-1, 8)
plt.ylim(-1, 8)

# Draw axes
plt.axvline(x=0, color= 'grey')
plt.axhline(y=0, color= 'grey')
plt.scatter(4, 7, marker='x', s=50)

# Draw the name of the vectors
plt.text(-.5, 2, r'$\vec{u}$', color = 'grey', size=18)
plt.text(.5, 4.5, r'$\vec{v}$', color = 'grey', size=18)
plt.text(2.5, 7, r'$\vec{v}$', color = 'grey', size=18)

plt.show()
plt.close()
```

```python
# Two equations: One Unknown 

# Define x and y values
x = np.arange(-10, 10)
y = 0.5*x + 1
y1 = -x + 4

# Create Graph
plt.figure()

# Plot Equations
plt.plot(x, y)
plt.plot(x, y1)

# Label Axes
plt.xlim(-2, 10)
plt.ylim(-2, 10)

# Draw Axes
plt.axvline(x=0, color= 'Grey')
plt.axhline(y=0, color= 'Grey')

plt.show()
plt.close()
```

```python
### 2 scaled vectors, solution = span

# Define scaled and unscaled vectors
u = [0,0,0.5,1]
u_bis = [u[2], u[3], u[2], u[3]]
v = [2*u[2],2*u[3],-1,1]
v_bis = [2*u[2]-1,2*u[3]+1,v[2],v[3]]

# Plot Arrows
plt.quiver([u[0], u_bis[0], v[0], v_bis[0]],
           [u[1], u_bis[1], v[1], v_bis[1]],
           [u[2], u_bis[2], v[2], v_bis[2]],
           [u[3], u_bis[3], v[3], v_bis[3]],
           angles='xy', scale_units='xy', scale=1, color=['Blue', 'Blue', 'Orange', 'Orange'])

# Label Axes
plt.xlim(-1.5, 2)
plt.ylim(-0.5, 4.5)

# Draw axes
plt.axvline(x=0, color='#A9A9A9')
plt.axhline(y=0, color='#A9A9A9')

# Mark end point of vectors
plt.scatter(-1,4,marker='x',s=50)

# Name vectors
plt.text(0, 0.5, r'$\vec{u}$', color='Blue', size=18)
plt.text(0.5, 1.5, r'$\vec{u}$', color='Blue', size=18)
plt.text(0.5, 2.7, r'$\vec{v}$', color='Orange', size=18)
plt.text(-0.8, 3, r'$\vec{v}$', color='Orange', size=18)

plt.show()
plt.close()
```

```python
### Parallel lines: No Solutions 

x = np.arange(-10, 10)
y = 2*x + 6
y1 = 2*x 

# Create figure and plot equations
plt.figure()
plt.plot(x, y)
plt.plot(x, y1)

# Label Axes 
plt.xlim(-2, 10)
plt.ylim(-2, 10)

# Draw Axis Lines
plt.axvline(x=0, color = 'Grey')
plt.axhline(y=0, color = 'Grey')

plt.show()
plt.close()
```

```python
### No Solution, Vectors that go in different directions 
### No unknowns = overdetermined system 

u = [0,0,2,2]
v = [0,0,-1,-1]

# Draw Arrows
plt.quiver([u[0], v[0]],
           [u[1], v[1]],
           [u[2], v[2]],
           [u[3], v[3]],
           angles='xy', scale_units='xy', scale=1, color=['Blue', 'Orange'])

# Label Axes
plt.xlim(-7, 3)
plt.ylim(-2, 3)

# Draw axes
plt.axvline(x = 0, color= 'Grey')
plt.axhline(y = 0, color= 'Grey')

# Mark Vector End Point and Label
plt.scatter(-6, 0, marker = 'x', s = 150)
plt.text(-6, 0.5, r'$b$', color='b', size=18)

plt.show()
plt.close()
```
